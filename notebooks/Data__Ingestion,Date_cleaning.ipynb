```python
import requests
import os
folder = "/content/drive/MyDrive/mutual-fund-dashboard/data/raw"
os.makedirs(folder, exist_ok=True)

schemes = {
    "SBI_Bluechip": 119551,
    "ICICI_Bluechip": 120503,
    "Nippon_Large_Cap": 118632,
    "Axis_Bluechip": 119092,
    "Kotak_Bluechip": 120841
}

for scheme_name, scheme_id in schemes.items():

    url = f"https://api.mfapi.in/mf/{scheme_id}"

    response = requests.get(url)

    if response.status_code == 200:

        data = response.json()

        nav_df = pd.DataFrame(data["data"])

        file_name = f"{scheme_name}.csv"
        file_path = os.path.join(folder, file_name)

        nav_df.to_csv(file_path, index=False)

        print(f" Saved {file_name}")
```

```python
import pandas as pd
ds=pd.read_csv("/content/drive/MyDrive/Google AI Studio/07_scheme_performance.csv")
```

```python
print(ds.columns)
print(ds.head())
```

```python
return_cols=['amfi_code', 'scheme_name', 'fund_house', 'category', 'plan',
       'return_1yr_pct', 'return_3yr_pct', 'return_5yr_pct',
       'benchmark_3yr_pct', 'alpha', 'beta', 'sharpe_ratio', 'sortino_ratio',
       'std_dev_ann_pct', 'max_drawdown_pct', 'aum_crore', 'expense_ratio_pct',
       'morningstar_rating', 'risk_grade']
```

```python
for col in return_cols:
     ds[col]=pd.to_numeric(ds[col],errors='coerce')
```

```python
invalid=ds[ds[return_cols].isna().any(axis=1)]
print(len(invalid))
invalid.head()
```

```python
ds[return_cols].isnull().sum()
```

```python
print(type(return_cols))
```

```python
ds.dropna(how='all')
ds.isnull().sum()
```

```python
expense_issue=ds[(ds["expense_ratio_pct"] < 0.1) |
  (ds["expense_ratio_pct"]>2.5)
]
print(len(expense_issue))
```

```python
for col in ["return_1yr_pct","return_3yr_pct","return_5yr_pct"]:
  print(col)
  print("Min=",ds[col].min())
  print("Max",ds[col].max())
```

```python
return_anomalies = ds[
    (ds["return_1yr_pct"] < -100) |
    (ds["return_1yr_pct"] > 100) |
    (ds["return_3yr_pct"] < -100) |
    (ds["return_3yr_pct"] > 100) |
    (ds["return_5yr_pct"] < -100) |
    (ds["return_5yr_pct"] > 100)
]

print(len(return_anomalies))
```

```python
ds[return_cols].describe()
```

```python
beta_issue=ds[
    (ds["beta"]<0)|
    (ds["beta"]>3)
]
print(len(beta_issue))
```

```python
sharpe_issue = ds[
    (ds["sharpe_ratio"] < -5) |
    (ds["sharpe_ratio"] > 10)
]

print( len(sharpe_issue))
```

```python
drawdown_issue = ds[
    (ds["max_drawdown_pct"] > 0) |
    (ds["max_drawdown_pct"] < -100)
]

print(len(drawdown_issue))
```

```python
ds["anomaly_flag"] = (
    (ds["expense_ratio_pct"] < 0.1) |
    (ds["expense_ratio_pct"] > 2.5) |
    (ds["beta"] < 0) |
    (ds["beta"] > 3) |
    (ds["max_drawdown_pct"] > 0) |
    (ds["max_drawdown_pct"] < -100)
)
```

```python
print(ds['anomaly_flag'].sum())
```
